### Linear Equations

An equation is an statement that equates to expressions. In mathematics, a linear equation is an equation that may be put in the form of $ax + by + c = z$. In this form the values a is called the *coefficients*, that are often real numbers.*x,y,z* are called the variables. The *c* is often used for constants. 

If the equation has only one undefined variabe then we call the it Linear equation in one variabe.These equations take the form $a x + b = 0$.

Similarly, equations with two undefined variabes are called Linear equation in two variabe. These equations take generally take the form $a x + by + c = 0$. 

We can also have equations with multiple variable called polynomials. What is a polynomial? the word "polynomial" just means many terms. And a polynomial is an algebraic expression, that contains constants, variables with coefficients, and exponentials with a positive exponent.

#### Slope-Intercept Form

A line can be defined by its *m* slope, and its y-intercept y0 (the y coordinate where is intersects the y-axis). 
The most common form is the slope-intercept equation of a straight line: 

\begin{align}
y & = mx+b \\
\end{align}

#### Plotting slope intercept form

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)

# https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
x = np.linspace(-10, 10, 1000)

# https://matplotlib.org/3.1.1/api/spines_api.html
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_position('center')
ax.spines['top'].set_position('center')

plt.plot(x, 2*x+1, '-r', label = 'y = 2x+1')
plt.plot(x, 2*x-3, '-b', label = 'y = 2x+1')

plt.legend(loc = 'upper left')
plt.show()
```

### Functions

Sometimes a equation is written as a function, with f(x) instead of y. The above slope intecept equation can be written as $f(x) = mx + b$. We see functions everywhere in our lives, anything that has a input, process and output is a function. There are other functions like $x2$ is a squaring function, $y3+1$ is also a function and in geometry, Sine, Cosine and Tangent are functions.

#### Vector Addition

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

v = np.array([2,4,6])

u = np.array([-5, -3, -2])

z = np.add(v,u)

fig = plt.figure()

ax = Axes3D(fig)

# x,y,z

ax.plot([0,0], [0,0], [-10, 10], 'k--')
ax.plot([0,0], [-10, 10], [0,0], 'k--')
ax.plot( [-10, 10], [0,0], [0,0],  'k--')

print (z)
ax.plot([0, v[0]],[0,v[1]],[0,v[2]])
ax.plot([0, u[0]],[0,u[1]],[0,u[2]])
ax.plot([0, z[0]],[0,z[1]],[0,z[2]])
```

#### Vector Multiplication

Vector Multiplication is of three types "Scalar product", Cross product and Dot product.

In an scalar product we multiply the number directly by a scalar quantity and our vector scales by that much value.

A dot product can be thought as direction growth on a vector when multiplied with another vector, the result gives us the effect on the original vector(positive, negative, or zero). Dot product is a scalar quantity. We calculate the dot product using this formula $a  b = |a|  |b|  cos()$

Cross Product is defined as a vector that is at right angles to the other two vectors . We calculate cross product using this formula $a  b = |a| |b| sin() n$. We leverage the "Right Hand Rule" to better understand the cross product.

#### Try it out

If you point your index finger along vector a, and point your middle finger along vector b: the cross product goes in the direction of your thumb. , 

```python
u = np.array([1,2,3])
v = np.array([5,2,4])

# Dot product

dotp = np.dot(u,v)

# Cross Product

crossp = np.cross(u,v)

print("Dot Product is : {}".format(dotp))

print("Cross Product is : {}".format(crossp))
```
